```python
punc = r'''!()[]{};:"\,<>./?@#$%^&*_~«»”—“…'''
```

```python
def remove_punctuation(array):
    for element in array:
      for ele in element:
        if ele in punc:
            element = element.replace(ele, " ")
    return element
```

```python
def label_to_author(label):
    return authors[int(label.replace("LABEL_", ""))]
```

```python
def to_lower(array):
  for element in array:
   return element.lower()
```

```python
def remove_digits(array):
    for element in array:
      for ele in element:
        if ele.isdigit():
            element.replace(ele,"")
    return element
```

```python
import time
from transformers import BertTokenizer, BertForSequenceClassification, TextClassificationPipeline
import numpy as np
```

```python
tokenizer = BertTokenizer.from_pretrained("bert-base-multilingual-cased")
model = BertForSequenceClassification.from_pretrained("/content/drive/MyDrive/ModelTakeTwo/checkpoint-73000")
```

```python
authors = ['Ataol Behramoğlu', 'Attila İlhan', 'Cahit Sıtkı Tarancı', 'Cahit Zarifoğlu',
 'Can Yücel', 'Cemal Safi', 'Cemal Süreya', 'Edip Cansever',
 'Mehmet Akif Ersoy', 'Nazım Hikmet Ran', 'Necip Fazıl Kısakürek',
 'Orhan Veli Kanık', 'Turgut Uyar', 'Özdemir Asaf', 'Ümit Yaşar Oğuzcan']
```

```python
pipe = TextClassificationPipeline(model=model, tokenizer=tokenizer,return_all_scores = True)
```

```python

text_path = input("Şiiri giriniz: ")
if text_path == "sahte":
  text_path = "/content/drive/MyDrive/Necip Fazıl Kısakürek (Sahte).txt"
elif text_path == "gerçek":
  text_path = "/content/drive/MyDrive/Necip Fazıl Kısakürek.txt"

print("")
poet = input("Şairi giriniz: ")
time.sleep(0.5)
print("")
print("Şairin doğruluğu test ediliyor... \n")


with open(text_path) as f:
    lines = np.array(f.readlines())

remove_punctuation(lines)
to_lower(lines)
remove_digits(lines)

for i in range(0,len(lines)):
  if not lines[i]:
    lines.pop(i)


data = []
for i in range(0,len(lines)):
  lines[i] = lines[i].replace("\n","")
  data.append(pipe(lines[i]))

label_to_score_mean = dict()

for predictions in data:
    for prediction in predictions[0]:
        prediction["label"] = label_to_author(prediction["label"])
        if prediction["label"] in label_to_score_mean:
            label_to_score_mean[prediction["label"]] += prediction["score"]
        else:
            label_to_score_mean.update({prediction["label"]: prediction["score"]})

for key, value in label_to_score_mean.items():
    label_to_score_mean[key] = round(value/len(data) *100,2)

predicted_author = list(label_to_score_mean.keys())[list(label_to_score_mean.values()).index(max(label_to_score_mean.values()))]
possibility = label_to_score_mean[predicted_author]

if poet.lower() == predicted_author.lower():
    print(f"Bu bilgi %{possibility} ihtimalle doğrudur.")
else:
  print("Bu bilgi yanlıştır.")
  print(f"Bu şiir %{possibility} ihtimalle {predicted_author} tarafından yazılmıştır.")

print("")

time.sleep(1)
desire = input("Bütün şairlerin olasılıklarını görmek ister misiniz(E/h): ")
desire = desire.lower()

if desire == "e":
  print("")
  for key, value in label_to_score_mean.items():
    print(f"Bu şiir %{value} ihtimalle {key} tarafından yazılmıştır.")
    time.sleep(0.25)
  
  print("")
else:
  print("")

time.sleep(0.5)
print("Program kapatılıyor...")


```
